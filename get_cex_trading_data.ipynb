```python
import ccxt
import pandas as pd
from time import sleep
from datetime import datetime, timedelta
from pprint import pprint
from const import symbols
```

```python
pprint(ccxt.exchanges)
```

```python
def limit(tf):
    return 288 if tf == '5m' else 288 * 5 if tf == '1m' else 1000
```

```python
# Vars 
date_start = '2022-01-01'
nb_jours = 30
tf = "5m"
```

```python
binance = ccxt.binance()

btc_ohlcv = []

for i in range(0, nb_jours):
    start = int((datetime.strptime(date_start, "%Y-%m-%d") +
                timedelta(days=i)).timestamp() * 1000)
    btc_ohlcv += binance.fetch_ohlcv(
        'BTC/BUSD', timeframe=tf, since=start, limit=limit(tf))
    sleep(1)
    
print(len(btc_ohlcv))
pprint(btc_ohlcv)
```

```python
df_btc = pd.DataFrame(btc_ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
print(df_btc)
```
